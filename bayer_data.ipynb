```python
bayern_loc_df = ppda_df[(~ppda_df['location'].isnull()) & (ppda_df['minute']<59) & (ppda_df['team']=='Bayern Munich')]
bayer_loc_df = ppda_df[(~ppda_df['location'].isnull()) & (ppda_df['minute']<64) & (ppda_df['team']=='Bayer Leverkusen')]
```

```python
bayern_loc_df['x'] = bayern_loc_df['location'].apply(lambda x:x[0])
bayern_loc_df['y'] = bayern_loc_df['location'].apply(lambda x:x[1])

bayer_loc_df['x'] = bayer_loc_df['location'].apply(lambda x:x[0])
bayer_loc_df['y'] = bayer_loc_df['location'].apply(lambda x:x[1])

```

```python
ppda_df[ppda_df['type']=='Substitution'][['minute','player','team']]
```

```python
bayern_avg_df = bayern_loc_df.groupby(['team','player'])[['x','y']].mean().reset_index()
bayer_avg_df = bayer_loc_df.groupby(['team','player'])[['x','y']].mean().reset_index()
```
